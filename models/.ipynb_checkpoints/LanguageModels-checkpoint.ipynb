```python
import re
import nltk  #Natural language processing tool-kit
from nltk.corpus import stopwords

temp = []
snow = nltk.stem.SnowballStemmer('english')
stop = set(stopwords.words('english'))

contractions = {
    "ain't": "am not",
    "aren't": "are not",
    "can't": "can not",
    "can't've": "can not have",
    "'cause": "because",
    "could've": "could have",
    "couldn't": "could not",
    "couldn't've": "could not have",
    "didn't": "did not",
    "doesn't": "does not",
    "don't": "do not",
    "hadn't": "had not",
    "hadn't've": "had not have",
    "hasn't": "has not",
    "haven't": "have not",
    "he'd": "he would",
    "he'd've": "he would have",
    "he'll": "he will",
    "he'll've": "he will have",
    "he's": "he is",
    "how'd": "how did",
    "how'd'y": "how do you",
    "how'll": "how will",
    "how's": "how is",
    "i'd": "i would",
    "i'd've": "i would have",
    "i'll": "i will",
    "i'll've": "i will have",
    "i'm": "i am",
    "i've": "i have",
    "isn't": "is not",
    "it'd": "it would",
    "it'd've": "it would have",
    "it'll": "it will",
    "it'll've": "it will have",
    "it's": "it is",
    "let's": "let us",
    "ma'am": "madam",
    "mayn't": "may not",
    "might've": "might have",
    "mightn't": "might not",
    "mightn't've": "might not have",
    "must've": "must have",
    "mustn't": "must not",
    "mustn't've": "must not have",
    "needn't": "need not",
    "needn't've": "need not have",
    "o'clock": "of the clock",
    "oughtn't": "ought not",
    "oughtn't've": "ought not have",
    "shan't": "shall not",
    "sha'n't": "shall not",
    "shan't've": "shall not have",
    "she'd": "she would",
    "she'd've": "she would have",
    "she'll": "she will",
    "she'll've": "she will have",
    "she's": "she is",
    "should've": "should have",
    "shouldn't": "should not",
    "shouldn't've": "should not have",
    "so've": "so have",
    "so's": "so as",
    "that'd": "that would",
    "that'd've": "that would have",
    "that's": "that is",
    "there'd": "there would",
    "there'd've": "there would have",
    "there's": "there is",
    "they'd": "they would",
    "they'd've": "they would have",
    "they'll": "they will",
    "they'll've": "they will have",
    "they're": "they are",
    "they've": "they have",
    "to've": "to have",
    "wasn't": "was not",
    "we'd": "we would",
    "we'd've": "we would have",
    "we'll": "we will",
    "we'll've": "we will have",
    "we're": "we are",
    "we've": "we have",
    "weren't": "were not",
    "what'll": "what will",
    "what'll've": "what will have",
    "what're": "what are",
    "what's": "what is",
    "what've": "what have",
    "when's": "when is",
    "when've": "when have",
    "where'd": "where did",
    "where's": "where is",
    "where've": "where have",
    "who'll": "who will",
    "who'll've": "who will have",
    "who's": "who is",
    "who've": "who have",
    "why's": "why is",
    "why've": "why have",
    "will've": "will have",
    "won't": "will not",
    "won't've": "will not have",
    "would've": "would have",
    "wouldn't": "would not",
    "wouldn't've": "would not have",
    "y'all": "you all",
    "y'all'd": "you all would",
    "y'all'd've": "you all would have",
    "y'all're": "you all are",
    "y'all've": "you all have",
    "you'd": "you would",
    "you'd've": "you would have",
    "you'll": "you will",
    "you'll've": "you will have",
    "you're": "you are",
    "you've": "you have"
}


def preprocess(doc):
    sentence = doc.lower()  # Converting to lowercase
    cleaner = re.compile('<.*?>')
    sentence = re.sub(cleaner, ' ', sentence)  #Removing HTML tags
    sentence = re.sub(r'[?|!|\'|"|#]', r'', sentence)
    sentence = re.sub(r'[.|,|)|(|\|/]', r' ', sentence)
    sentence = re.sub(r'[^a-zA-Z]', ' ', sentence)

    q_decontracted = []

    for word in sentence.split():
        if word in contractions:
            word = contractions[word]

        q_decontracted.append(word)

    q = ' '.join(q_decontracted)
    q = q.replace("'ve", " have")
    q = q.replace("n't", " not")
    q = q.replace("'re", " are")
    q = q.replace("'ll", " will")

    sentence = q

    sentence = " ".join([snow.stem(word) for word in sentence.split() if word not in stopwords.words('english')])

    return sentence
```

```python

def precision_at_k(gold, retrieved, k):
    ''' returns avarage of precision at rank k from a gold and retrieved data.
    '''

    sum_precision_at_k = 0
    count_q = 0
    for q in retrieved:

        top_ten = (sorted(retrieved[q], key=lambda item: item[1], reverse=True)[
                   :10])

        gold_qs = list(gold.loc[gold['query_id'] == q].doc_id)

        if len(gold_qs) != 0:
            count_q = count_q + 1
        else:
            continue

        sum_correct = 0
        for idx, (result_id, result_text) in enumerate(top_ten):
            if result_id in gold_qs:
                sum_correct = sum_correct + 1
            if idx == k - 1:
                sum_precision_at_k = sum_precision_at_k + (sum_correct / k)
                break

    return sum_precision_at_k / count_q
```

```python
def mean_average_precision(gold, retrieved):
    ''' returns avarage of avarage precisions from a gold and retrieved data.
    '''

    sum_average_precision = 0
    count_q = 0
    for q in retrieved:

        top_ten = (sorted(retrieved[q], key=lambda item: item[1], reverse=True)[
                   :10])
        gold_qs = list(gold.loc[gold['query_id'] == q].doc_id)

        if len(gold_qs) != 0:
            count_q = count_q + 1
        else:
            continue

        sum_precision = 0
        sum_correct = 0

        for idx, (result_id, result_text) in enumerate(top_ten):
            if result_id in gold_qs:
                sum_correct = sum_correct + 1
                sum_precision = sum_precision + (sum_correct / (idx + 1))
        if sum_correct != 0:
            sum_average_precision = sum_average_precision + \
                                    (sum_precision / sum_correct)

    return sum_average_precision / count_q
```

```python
def mean_reciprocal_rank(gold, retrieved):
    ''' returns avarage of reciprocal ranks from a gold and retrieved data.
    for all evaluation methods, retrieved dictionary must have queries as keys, and list of (doc id , p) as values.
    '''

    sum_reciprocal_rank = 0
    count_q = 0
    for q in retrieved:

        # get top ten values from retrieved dictionary for every query
        top_ten = (
            sorted(retrieved[q], key=lambda item: item[1], reverse=True)[:10])

        # get gold data for every query from gold dictionary
        gold_qs = list(gold.loc[gold['query_id'] == q].doc_id)

        # ignore queries that don't have any gold data with is_duplicate set to 1
        if len(gold_qs) != 0:
            count_q = count_q + 1
        else:
            continue

        reciprocal_rank = 0
        # find the first relevant retrieved doc and compute reciprocal rank and add it to sum
        for idx, (result_id, result_text) in enumerate(top_ten):
            if result_id in gold_qs:
                reciprocal_rank = 1 / (idx + 1)
                sum_reciprocal_rank = sum_reciprocal_rank + reciprocal_rank
                break

    # avarage of reciprocal ranks
    return sum_reciprocal_rank / count_q
```

## Unigram

```python
def make_vocab_vector(docs, column_name):
    ''' for a dictionary of questions, returns a dictionary of uniqued words and initializes the term frequency and corpus frequency
    term frequency is in form of a dictionary with keys being doc ids and value being the term frequency of the mentioned word in that doc.
    {"hello" : {"tfs": {}, "cf": 0}}
    '''
    vocab = {}

    for ind in docs.index:
        words = docs[column_name][ind].split()
        for word in words:
            vocab[word] = {}
            vocab[word]["tf"] = {}
            vocab[word]['cf'] = 0

    return vocab
```

```python
def make_tf_cf_vocab_vector(docs, column_id, column_name):
    vocab_vector = make_vocab_vector(docs, column_name)

    for ind in docs.index:
        terms = docs[column_name][ind].split()
        doc_id = docs[column_id][ind]

        for term in terms:
            if ind in vocab_vector[term]['tf']:
                vocab_vector[term]['tf'][doc_id] = vocab_vector[term]['tf'][doc_id] + 1
                vocab_vector[term]['cf'] = vocab_vector[term]['cf'] + 1
            else:
                vocab_vector[term]['tf'][doc_id] = 1
                vocab_vector[term]['cf'] = vocab_vector[term]['cf'] + 1

    all_corpus_tokens = 0
    for word in vocab_vector:
        all_corpus_tokens = all_corpus_tokens + vocab_vector[word]["cf"]

    return all_corpus_tokens, vocab_vector
```

```python
def compute_jelinek_mercer_smoothed(vocab_vector, word, doc_id, doc_len, c, l):
    ''' returns the jelinek_mercer smoothed probability for a word with the condition of a document'''

    if word in vocab_vector:
        if doc_id in vocab_vector[word]['tf']:
            # if the doc has the word...

            p = (l * vocab_vector[word]['tf'][doc_id] / doc_len) + ((1 - l) * (vocab_vector[word]['cf'] / c))
        else:
            # if the doc has not the word so the TF part of equation is zero.
            p = (1 - l) * (vocab_vector[word]['cf'] / c)
    else:
        # if the word is not in vocab vector
        return 1

    return p
```

```python
all_corpus_tokens, vocab_vector = make_tf_cf_vocab_vector(docs, 'doc_id', 'processed_document')
```

```python
def compute_unigram(vocab_vector, queries, docs, c, landa):
    result = {}
    for q_ind in queries.index:
        q_id = queries['query_id'][q_ind]
        q_doc = queries['processed_query'][q_ind]
        result[q_id] = []

        q_words = q_doc.split()

        for d_ind in docs.index:

            d_id = docs['doc_id'][d_ind]
            d_doc = docs['processed_document'][d_ind]
            p_doc = 1

            for word in q_words:
                # preproccessing and getting words of doc
                doc_words = d_doc.split()
                doc_len = len(doc_words)

                # compute dirichlet smoothed p and multiply it by previous ps
                p = compute_jelinek_mercer_smoothed(
                    vocab_vector, word, d_id, doc_len, c, landa)
                p_doc = p_doc * p

            result[q_id].append((d_id, p_doc))

    return result
```

```python
unigram_result = compute_unigram(vocab_vector, queries, docs, all_corpus_tokens, 0.1)
```

```python
lambda_values = [0.001, 0.01, 0.1, 0.3, 0.5, 0.7]

for l in lambda_values:
    res = compute_unigram(vocab_vector, queries, docs, all_corpus_tokens, l)
    mrr = mean_reciprocal_rank(qrels, res)
    map = mean_average_precision(qrels, res)
    p_at_5 = precision_at_k(qrels, res, 5)
    p_at_10 = precision_at_k(qrels, res, 10)
    print("landa={} => MRR:{:.2f} --- MAP:{:.2f} --- P@5:{:.2f} --- P@10:{:.2f}\n".format(l, mrr, map, p_at_5, p_at_10))
```

## Bigram

```python
def compute_smoothed_bigram_p(vocab_vector, word_i, word_i_1, doc_id, doc_len, doc_words, c, landa1, landa2):
    ''' returns smoothed bigram probability for word i, if word i-1 and doc.
    doc_words is list of doc words and is needed for computing bigrams of document.
    landa1 and landa2 are parameters.'''

    p = 0
    p_smoothed_uni = 0
    count = 0

    # if both words are in vocab vector
    if word_i in vocab_vector and word_i_1 in vocab_vector:

        # if both words are in doc
        if doc_id in vocab_vector[word_i]['tf'] and doc_id in vocab_vector[word_i_1]['tf']:

            # compute count of this bigram in doc
            for idx, word in enumerate(doc_words):
                if idx != len(doc_words) - 1:
                    if word == word_i_1 and doc_words[idx + 1] == word_i:
                        count = count + 1

            # first part of smoothed bigram formula
            p = count / vocab_vector[word_i_1]['tf'][doc_id]

    # second part of smoothed bigram formula

    if word_i in vocab_vector:
        if doc_id in vocab_vector[word_i]['tf']:
            # if the doc has the word...

            p_smoothed_uni = (landa2 * vocab_vector[word_i]['tf'][doc_id] / doc_len) + (
                    (1 - landa1 - landa2) * (vocab_vector[word_i]['cf'] / c))
        else:
            # if the doc has not the word so the TF part of equation is zero.
            p_smoothed_uni = (1 - landa1 - landa2) * (vocab_vector[word_i]['cf'] / c)

    # applying landa
    return (landa1 * p) + p_smoothed_uni
```

```python
def compute_bigram_model(vocab_vector, queries, docs, c, landa1, landa2):
    ''' returns result dictionary for bigram model of a dict of queries and docs.
    '''

    result = {}
    for q_ind in queries.index:
        q_id = queries['query_id'][q_ind]
        q_doc = queries['processed_query'][q_ind]
        result[q_id] = []

        q_words = q_doc.split()

        for d_ind in docs.index:

            d_id = docs['doc_id'][d_ind]
            d_doc = docs['processed_document'][d_ind]

            p2 = 1
            for idx, word in enumerate(q_words):

                doc_words = d_doc.split()
                doc_len = len(doc_words)

                # compute unigram p for first query word
                if idx == 0:
                    p2 = p2 * compute_jelinek_mercer_smoothed(vocab_vector, word, d_id, doc_len, c, landa1)

                # compute bigram p for the rest of query words
                else:
                    p2 = p2 * compute_smoothed_bigram_p(vocab_vector, word, q_words[idx - 1], d_id, doc_len, doc_words,
                                                        c, landa1, landa2)

            result[q_id].append((d_id, p2))

    return result
```

```python
compute_bigram_model(vocab_vector, queries, docs, all_corpus_tokens, 0.2, 0.3)
```

```python
lambda_values2 = [0.001, 0.01, 0.1, 0.3, 0.5, 0.7]

for l in lambda_values2:
    res = compute_bigram_model(vocab_vector, queries, docs, all_corpus_tokens, 0.3, l)
    mrr = mean_reciprocal_rank(qrels, res)
    map = mean_average_precision(qrels, res)
    p_at_5 = precision_at_k(qrels, res, 5)
    p_at_10 = precision_at_k(qrels, res, 10)
    print("landa1={} landa2={} => MRR:{:.2f} --- MAP:{:.2f} --- P@5:{:.2f} --- P@10:{:.2f}\n".format(0.3, l, mrr, map,
                                                                                                     p_at_5, p_at_10))
    
```
